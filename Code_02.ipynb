# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
# -- Variáveis das imagens --
imagem1 = Image.open('/content/drive/My Drive/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')
imagem2 = Image.open('/content/drive/My Drive/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem3 = Image.open('/content/drive/My Drive/bailey-zindel-NRQV-hBF10M-unsplash.jpg')


# -- Plotando numa grade (1x3) --
plt.figure(figsize = (20, 10))

plt.subplot(131)
plt.axis('off')
plt.title('Imagem 1')
plt.imshow(imagem1)

plt.subplot(132)
plt.axis('off')
plt.title('Imagem 2')
plt.imshow(imagem2)

plt.subplot(133)
plt.axis('off')
plt.title('Imagem 3')
plt.imshow(imagem3)

plt.show()


# -- Arrays das imagens e cada cor --
imagem1_arr = np.array(imagem1)
imagem2_arr = np.array(imagem2)
imagem3_arr = np.array(imagem3)

red1 = imagem1_arr[:,:,0]
green1 = imagem1_arr[:,:,1]
blue1 = imagem1_arr[:,:,2]

red2 = imagem2_arr[:,:,0]
green2 = imagem2_arr[:,:,1]
blue2 = imagem2_arr[:,:,2]

red3 = imagem3_arr[:,:,0]
green3 = imagem3_arr[:,:,1]
blue3 = imagem3_arr[:,:,2]


# -- Mostrando histogramas --
plt.figure(figsize=(20,15))

plt.subplot(331)
plt.hist(red1.ravel(), color='red')
plt.subplot(332)
plt.hist(red2.ravel(), color='red')
plt.subplot(333)
plt.hist(red3.ravel(), color='red')

plt.subplot(334)
plt.hist(green1.ravel(), color='green')
plt.subplot(335)
plt.hist(green2.ravel(), color='green')
plt.subplot(336)
plt.hist(green3.ravel(), color='green')

plt.subplot(337)
plt.hist(blue1.ravel(), color='blue')
plt.subplot(338)
plt.hist(blue2.ravel(), color='blue')
plt.subplot(339)
plt.hist(blue3.ravel(), color='blue')

plt.show()
```

Olhando para os gráficos gerados, discorram comparando os resultados.



> Na primeira imagem há altos valores de vermelho, baixos valores de azul e alguns tons de verde.

> Na segunda imagem há valores médios de vermelho e verde e baixos valores de azul.

> Na terceira imagem há altos valores de azul e verde, e altos e baixos valores de vermelho.
